Justificación para no imputar outliers
dado el contexto del estudio y la naturaleza de las variables, no se realizará imputación de outliers porque:

1. Los valores extremos reflejan la heterogeneidad real del sector agropecuario peruano.
Las unidades agropecuarias varían significativamente en tamaño, tipo de producción y nivel de tecnificación. Por ejemplo, algunas parcelas son explotaciones familiares pequeñas, mientras que otras corresponden a grandes empresas agrícolas que naturalmente concentran más trabajadores, mayor superficie y mayor producción. Los "outliers" en este contexto no representan errores, sino diversidad estructural del sector.

2. La imputación basada en reglas estadísticas (como el IQR) podría distorsionar patrones legítimos.
Truncar o reemplazar valores extremos por límites artificiales podría subestimar la magnitud real de producción, gastos o empleo en unidades grandes, afectando el análisis descriptivo y cualquier modelo predictivo o multivariado que dependa de esta variabilidad.

3. Las variables clave (como superficie trabajada, número de trabajadores, gastos y cantidades producidas) están estrechamente interrelacionadas.
Los valores altos en una variable (ej. superficie sembrada) probablemente se correspondan con valores altos en otras (ej. cantidad de trabajadores o gasto en insumos). Imputar algunos outliers pero no otros rompería estas relaciones naturales dentro de las unidades de análisis.

4. El instrumento de recolección (cuestionario digital y entrevistas directas) y el diseño muestral validado por el INEI brindan alta confiabilidad a los datos recolectados, por lo que los valores extremos tienen alta probabilidad de ser reales y no errores de digitación.

5. Preservar la distribución original es crucial para entender la estructura productiva del agro peruano, especialmente si se busca analizar desigualdades, economías de escala, o diferencias por región, tipo de cultivo, o acceso a servicios.

**Justificación del uso de RobustScaler**
En el presente análisis, se ha optado por utilizar el método de normalización RobustScaler debido a la naturaleza heterogénea de los datos del sector agropecuario peruano. A continuación, se detallan los fundamentos técnicos y contextuales que motivan esta elección:

1. Presencia de valores extremos reales (no errores):
Los datos presentan una amplia variabilidad en variables como superficie trabajada, cantidad cosechada, gasto en insumos, y número de trabajadores. Estos valores extremos no son errores de digitación ni outliers espurios, sino que reflejan la diversidad estructural del agro peruano, que abarca desde pequeñas explotaciones familiares hasta grandes agroindustrias.

2. Conservación de la distribución original:
El uso de técnicas tradicionales como MinMaxScaler o StandardScaler podría distorsionar la distribución original de las variables debido a su sensibilidad a los outliers. En cambio, RobustScaler utiliza la mediana y el rango intercuartílico (IQR), lo que permite reducir la influencia de los valores extremos sin eliminarlos ni truncarlos.

3. Necesidad de preservar relaciones multivariadas reales:
Las variables agropecuarias están fuertemente correlacionadas (por ejemplo, mayor superficie suele implicar mayores costos, producción y mano de obra). Imputar o truncar outliers rompería estas relaciones naturales, mientras que RobustScaler mantiene la estructura relativa entre observaciones.

4. Enfoque exploratorio con clustering:
Dado que el análisis posterior incluye técnicas de agrupamiento como clustering, que dependen de métricas de distancia, es crucial que todas las variables estén escaladas en un rango comparable, sin que los valores extremos dominen la agrupación. RobustScaler es particularmente adecuado para este tipo de análisis, ya que permite resaltar patrones globales sin perder la información aportada por la heterogeneidad.
